```python
import re
from Bio import SeqIO

SEGMENTS = ["ha", "na", "mp", "pb1", "pa", "pb2", "np", "ns"]

path_to_fasta_files = "data/"

# File paths
InFileName = "config/exclude_strains.txt"

# Use a set to collect unique sequence IDs
unique_ids = set()

# Reads in a fasta file, and iterates through every sequence in it
for gene in SEGMENTS: 
    count = 0
    print(gene)
    input_fasta = path_to_fasta_files + "h5nx_" + gene + ".fasta"
     
    # seq.id = the name
    # seq.seq = the sequence itself 
    for seq in SeqIO.parse(input_fasta, "fasta"): 
        sequence = str(seq.seq)  # seq.seq is an object - this converts it to a string 
        sequence_id = str(seq.id) # seq.id is an object - this converts it to a string

        # Check if the sequence ends with the gene name
        if sequence.endswith(gene): 
            unique_ids.add(seq.id)  # Add the sequence ID to the set
            count += 1  # Increment count for this condition

        # Check if the gene is present within the sequence
        elif gene in sequence: 
            unique_ids.add(seq.id)  # Add the sequence ID to the set
            count += 1  # Increment count for this condition

        # Check if the sequence ID starts with 'r' followed by digits, OK but this doesn't exclude all of the Laboratoryderived seqs
        elif re.match(r'^r\d+', sequence_id):  # Using regular expression
            unique_ids.add(sequence_id)  # Add the sequence ID to the set
            count += 1  # Increment count for this condition

        else: 
            pass

    print(f"Total sequences for {gene}: {count}")

# Append the unique sequence IDs to the output file
with open(InFileName, 'a') as InFile:  # Open file in append mode to keep previous data
    for seq_id in unique_ids:
        InFile.write(seq_id + "\n")

print(f"Total unique sequences appended to {InFileName}: {len(unique_ids)}")
```
